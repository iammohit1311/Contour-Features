```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('/home/mohit.srivastava@Brightlifecare.local/Pictures/Screenshots/t2.png')
img_blured = cv2.GaussianBlur(img, (5, 5), 2)
img_edges = cv2.Canny(img_blured, 45, 140)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (2,2))
img_refined = cv2.morphologyEx(img_edges, cv2.MORPH_DILATE, kernel=kernel, iterations=3)

plt.imshow(img_refined, cmap="gray")
```

```python
# Find all contours in the image
contours, hierarchy = cv2.findContours(img_refined, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Display total contours found
print("Number of contours found = {}".format(len(contours)))
```

```python
# Function to display contour areas
def findArea(contours):
    areas = []
    for contour in contours:
        cont_area = cv2.contourArea(contour)
        areas.append(cont_area)
    return areas

sorted_contours_by_area = sorted(contours, key=cv2.contourArea, reverse=True)
print("Contour areas after sorting: ", findArea(sorted_contours_by_area))
```

```python
# Iterating over each contour and drawing it
for sc in sorted_contours_by_area:
    cv2.drawContours(img_refined, [sc], -1, (0, 0, 255), 3)
    cv2.waitKey(0)
    cv2.imshow("Highlight: ", img_refined)
    
cv2.waitKey(0)
cv2.destroyAllWindows()
```
